```python
class stpeter:
    '''
    The St. Petersburg game is a simple, coin-tossing game that first appeared in 1738 in a
    memoir by Daniel Bernoulli.
    
    The game ends when the coin shows heads for the first time. Formally, a single trial in 
    the St. Petersburg game consists of tossing a true coin until it falls as heads. 
    If it falls as heads at the nth throw,the gambler receives 2^ùëõ coins.
    
    (from The St. Petersburg paradox: an experimental solution by 
    Sergio Da Silva and Raul Matsushita)
    
    '''
    def __init__(self):
        self.lst = []
        self.n = int
        self.total = []
        self.suma = int
        self.i = 0
        self.j = 0
        self.media = []
        self.conta = int
        
    def cointoss(self):
        
        from random import randrange
        
               
        self.n = int(input('Insert the number of trials: '))
    
        # looping for i trials
        for self.i in range(self.n):
            self.coin = 0
            self.i += 1
            self.x = 0
            #print('Trial:', self.i)
            #print(' ')
            #print(' ')
            
            # looping to x turns in a trial (until you get a Head)
            # randrange = 0 Tail
            # randrange = 1 Head
            while self.coin != 1:
                self.coin = randrange(0,2)
                if self.coin == 0:
                    #print('Coin:',self.coin, 'Tail')
                    self.x +=1
                else:
                    #print('Coin', self.coin, 'Head')
                    self.x +=1
                    #print('Turn x : ', self.x)
                    #print(' ')
                    self.win = 2**(self.x)
                    self.lst.append(self.win)
    
    def totalwin(self):
        self.total.append(self.lst[0])
                
        for self.j in range(0,self.n-1):
            self.suma = self.total[self.j] + self.lst[self.j+1]
            self.total.append(self.suma)
        
    def average(self):
        self.k = 0

        for self.k in range(0,self.n):
            self.conta = self.total[self.k] / (self.k+1)
            self.media.append(self.conta)
   
    def visualize(self):
        import matplotlib.pyplot as plt
        %matplotlib inline
        
        self.fig = plt.figure()
        self.ax = self.fig.add_axes([0,0,1,1])
        self.ax.plot(range(0,self.n),self.media)
        self.ax.set_xlabel('Number of trials')
        self.ax.set_ylabel('Average')
        self.ax.set_title('St Peterburg Paradox')
        
    def lists_stpeter(self):
        print('Winnings: ',self.lst)
        print('Total: ', self.total)
        print('Average: ', self.media)
```

```python
# Function to play the class 
def play_peterburg():
    on = True

    while on == True:
        par = stpeter()
        par.cointoss()
        par.totalwin()
        par.average()
        par.visualize()

        on =False
```
